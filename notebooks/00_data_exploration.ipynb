




# 00 — Data Exploration: CUB-200-2011

Análisis exploratorio del dataset: distribución de clases, estadísticas de imágenes,
visualización de muestras y validación del pipeline de transforms.

```python
import sys
sys.path.insert(0, '../src')

import yaml
import torch
import matplotlib.pyplot as plt
from pathlib import Path
from collections import Counter

from geoquant.data.dataset import CUBDataset
from geoquant.utils.reproducibility import seed_everything

seed_everything(42)

with open('../configs/config.yaml') as f:
    config = yaml.safe_load(f)
```

```python
dataset = CUBDataset(root=config['data']['raw_dir'], image_size=224)
train_ds = dataset.get_split('train')
val_ds = dataset.get_split('test')

print(f'Train: {len(train_ds)} imágenes | Val: {len(val_ds)} imágenes')
print(f'Clases: {len(train_ds.classes)}')
```

```python
# Distribución de clases
label_counts = Counter([label for _, label in train_ds.samples])
plt.figure(figsize=(16, 3))
plt.bar(range(len(label_counts)), sorted(label_counts.values()))
plt.xlabel('Clase')
plt.ylabel('N muestras')
plt.title('Distribución de clases — CUB-200-2011 (train)')
plt.tight_layout()
plt.show()
```
